# Question 5
### 1.Like posts

```python
def to_like(numberofposts, word = 'None'):
    
    if word != 'None':            #checking input for other user
        searchandopen(word)
        time.sleep(5)
    ## Automating number of scrolls as per requirement

    n = numberofposts//3         #each row has 3 posts, so getting the number of rows for the required number of posts
    path = '//div[contains(@class, "Nnq7C")]'
    while True:
        x = driver.find_elements_by_xpath(path)
        if len(x)>n:
            break
        driver.execute_script('window.scrollBy(0,1000);') #scrolling down
        time.sleep(2)   
       
    
    ## Liking a post
    
    x = driver.find_elements_by_xpath('//div[contains(@class, "Nnq7C")]//div/a') #getting all post elements
    x = x[:numberofposts] #selecting the required number of posts
    a = 1
    for i in x:
        i.click() #opening the post
        WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.CLASS_NAME, 'fr66n')))
        path = "//span[contains(@class, 'fr66n')]/button/div/span/*[name()='svg']"
        WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, path)))
        like = driver.find_element_by_xpath(path)
        if like.get_attribute('aria-label') == 'Like':     #checking if it's already liked
            like.click()                            #liking the post 
        else:
            print('Post number',a,'is Aleady Liked')   #printing already liked since the post is liked.
        a+=1
        driver.find_element_by_xpath("//*[name()='svg'][@aria-label='Close']").click()   #closing the post     
    driver.execute_script('window.scrollTo(0,-document.body.scrollHeight);') #scrolling back to the top
    pass
```

```python
# If you want to like posts of the current user, don't give any input.
# If you want to like posts of  any other user, input the name of the user.

to_like(30,'dilsefoodie')
```

### 1.List Of Followers

```python
def list_of_followers(Numberoffollowers='all',word = 'None' ):
    
    if word != 'None': #checking input for other user
        searchandopen(word)
        time.sleep(5)
        
    ans = []
    #Opening the list of followers
    driver.find_element_by_xpath('//*[@id="react-root"]/section/main/div/header/section/ul/li[2]/a').click() 
    path="/html/body/div[5]/div/div/div[2]/ul/div//li"
    WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, path)))
    
    # for complete list we need the total number of followers and the below if statement code gives us that
    if Numberoffollowers=='all':
        str_num = driver.find_element_by_xpath('//*[@id="react-root"]/section/main/div/header/section/ul/li[2]/a').text.split()[0]
        if str_num[-1]=='k':
            Numberoffollowers = int(float(str_num[:-1])*1000)
        elif str_num[-1]=='M': 
            Numberoffollowers = int(float(str_num[:-1]*(10**6)))
        else:
            Numberoffollowers = int(float(str_num))
            
    n = len(driver.find_elements_by_xpath('/html/body/div[5]/div/div/div[2]/ul/div//li')) #number of followers that have loaded till now
    list_ = driver.find_elements_by_xpath(path) 
    
    while n<Numberoffollowers:         # below code scrolls till we get the desired number 
        WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, path)))
        list_ = driver.find_elements_by_xpath(path) 
        try:                        # scrolling to the last element in the list
            driver.execute_script("arguments[0].scrollIntoView();",list_[-1])
            time.sleep(2)
            WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, path)))
        except StaleElementReferenceException:
            continue
        n = len(driver.find_elements_by_xpath(path))
    
    for i in list_[:Numberoffollowers]:   #getting the list of desired number of followers 
        ans.append(i.text.split()[0])
    driver.find_element_by_xpath("//*[name()='svg'][@aria-label='Close']").click()   #closing the pop-up
    return ans
```

```python
# If you want to a list of followers of the current user, don't give any input.
# If you want to a list of followers of  any other user, input the name of the user.
# If you want to get a complete list, do not input a number 

ans = list_of_followers(500,'sodelhi')
ans
```

### 2.Listing people whom I follow but who don't follow me.

```python
def not_following(Numberoffollowers='all',word='None'):
    
    if word != 'None':
        searchandopen(word)
        time.sleep(5)
        
    ans = []
    page_followers_that_i_follow = []
    path = '//*[@id="react-root"]/section/main/div/header/section/ul/li[2]/a' 
    WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, path)))
    driver.find_element_by_xpath(path).click()
    path="/html/body/div[5]/div/div/div[2]/ul/div//li"
    WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, path)))
    
    if Numberoffollowers=='all':
        str_num = driver.find_element_by_xpath('//*[@id="react-root"]/section/main/div/header/section/ul/li[2]/a').text.split()[0]
        if str_num[-1]=='k':
            Numberoffollowers = int(float(str_num[:-1])*1000)
        elif str_num[-1]=='M': 
            Numberoffollowers = int(float(str_num[:-1]*(10**6)))
        else:
            Numberoffollowers = int(float(str_num))
   
    n = len(driver.find_elements_by_xpath('/html/body/div[5]/div/div/div[2]/ul/div//li'))
    list_ = driver.find_elements_by_xpath(path) 
    
    while True:
        if n > Numberoffollowers:
            break
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, path)))
        list_ = driver.find_elements_by_xpath(path) 
        try:
            driver.execute_script("arguments[0].scrollIntoView();",list_[-1])
            time.sleep(2)
            WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, path)))
        except StaleElementReferenceException:
            continue
        n = len(driver.find_elements_by_xpath(path))
    ## till here the code above is same as the previous code but the code below gives the list of people that I follow and who also follow the page 
    
    for i in list_[:Numberoffollowers]:
        if 'Following' in i.text: #checking if I follow them
            page_followers_that_i_follow.append(i.text.split('\n')[0])
    driver.find_element_by_xpath("//*[name()='svg'][@aria-label='Close']").click() #closing the pop-up
    
    time.sleep(3)
    
    path = '//*[@id="react-root"]/section/nav/div[2]/div/div/div[3]/div/div[5]/span'
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, path)))
    driver.find_element_by_xpath(path).click() # navigating to my profile
    
    time.sleep(3)
    
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//a[@class="-qQT3"]')))
    driver.find_element_by_xpath('//a[@class="-qQT3"]').click()

    time.sleep(3)

    my_followers = list_of_followers() # getting a list of people who follow me
    
    for i in page_followers_that_i_follow: #comparing to find the people who don't follow me
        if i not in my_followers:
            ans.append(i)
    if len(ans) == 0:
        return 'None'
    else:
        return ans
```

```python
def story(word='None'):
    if word!='None': #checking input for other user
        searchandopen(word)    
    path = '//*[@id="react-root"]/section/main/div/header/div/div/canvas'
    try: #checking if the user has added story
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, path)))
        height = int(driver.find_element_by_xpath(path).get_attribute('height'))
        if  height == 168: #checking if the story was seen 
            print('New/Unseen Story!')
            path = '//*[@id="react-root"]/section/main/div/header/div/div'
            driver.find_element_by_xpath(path).click()
        elif height == 166: #checking if the story was not seen
            print('You have already seen the story.')
    except NoSuchElementException:
        print('The user has no story!') 
```
